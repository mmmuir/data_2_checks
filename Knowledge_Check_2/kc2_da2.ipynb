```python
import pandas as pd
import matplotlib as plt
```

```python
# Read csv into dataframe, concat date and time columns into single datetime64 column
df = pd.read_csv('LMPD_STOP_DATA_2022.csv', parse_dates=[['ACTIVITY_DATE', 'ACTIVITY_TIME']])
df.head()
```

```python
# Correct spelling errors and shorten column names.
df = df.rename(columns={'WAS_VEHCILE_SEARCHED': 'WAS_VEHICLE_SEARCHED', 'ACTIVITY_DATE_ACTIVITY_TIME': 'DATETIME', 'CITATION_CONTROL_NUMBER': 'CITATION_NUMBER', 'ACTIVITY_LOCATION': 'LOCATION', 'ACTIVITY_DIVISION': 'DIVISION', 'ACTIVITY_BEAT': 'BEAT'})
```

```python
# Convert column from str to bool
df.loc[df.WAS_VEHICLE_SEARCHED == 'YES', 'WAS_VEHICLE_SEARCHED']=bool(1)
df.loc[(df.WAS_VEHICLE_SEARCHED == 'NO'), 'WAS_VEHICLE_SEARCHED']=bool(0)
```

```python
df.head()
```

```python
# Find null values
df.isnull().sum()
```

```python
# Drop null values
df.REASON_FOR_SEARCH = df.REASON_FOR_SEARCH.fillna(0)
df.DRIVER_RACE = df.DRIVER_RACE.fillna('UNKNOWN')
```

```python
# Show percentage of pulled-over drivers whose vehicles were searched, broken down by race.
pd.crosstab(df['DRIVER_RACE'], df['WAS_VEHICLE_SEARCHED']).apply(lambda r: r/r.sum() * 100, axis=1)
```
